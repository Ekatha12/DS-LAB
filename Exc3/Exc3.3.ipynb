```python
import matplotlib.pyplot as plt
colors=[]
percentage=[]
f=open('exc3.3.txt','r')
for row in f:
    row=row.split(' ')
    colors.append(row[0])
    percentage.append(int(row[1]))
plt.xlabel("Colors")
plt.ylabel("Percentage")
plt.plot(colors,percentage,marker='*',mfc='r',color='g')
plt.title("Color Percentage")
plt.show()
```
